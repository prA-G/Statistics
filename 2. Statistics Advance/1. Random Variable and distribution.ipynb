# *Random Variable*

*A **Random Variable** is a variable whose values depend on the outcomes of a random experiment.  
It assigns a **numerical value** to each possible outcome of a random process.*

A set of possible values from a random experiments.

### Example 1: Tossing a coin

- Experiment is random, outsomes will be random
- Either Head or Tail
- Quantify these random values

### Example 2: Throwing a Dice

- From 1 to 6 any face comes randomly.
- No. of possible outcomes can be 1/6
- Outcomes will be random.

## Types of Random Variables

1. **Discrete Random Variable**  
   - Takes **countable values** (finite or infinite but countable).  
   - Example: Number of heads when tossing 3 coins â†’ {0, 1, 2, 3}

2. **Continuous Random Variable**  
   - Takes **uncountable values** from an interval.  
   - Example: Height of students in a class (can be any real value in a range).

# **Discrete Random Variable**  

A **Discrete Random Variable (DRV)** is a type of random variable that can take only **countable values** (finite or infinite but countable).  

### Examples
- Number of heads when tossing 3 coins â†’ {0, 1, 2, 3}  
- Number of students present in class â†’ {0, 1, 2, â€¦, n}  
- Rolling a die â†’ {1, 2, 3, 4, 5, 6}  


# **Continuous Random Variable**

A **Continuous Random Variable (CRV)** is a type of random variable that can take **uncountably infinite values** in an interval.  
It is described using a **Probability Density Function (PDF)** instead of a PMF.


### Examples
- Height of students in a class (e.g., 150.3 cm, 171.8 cm, â€¦)  
- Time taken to finish a task (can be 12.35 sec, 12.351 sec, â€¦)  
- Temperature on a given day  

# Probability Distribution Function and Its Types

A **Probability Distribution Function** is a general term that explains how probabilities are assigned to the values of a random variable.  

There are **two types** depending on the kind of random variable:


## 1. Probability Mass Function (PMF)
- Used for **Discrete Random Variables**.  
- Gives the probability that the variable is **exactly equal** to some value.  

### Properties
1. \(0 \leq P(X=x) \leq 1\)  
2. \(\sum P(X=x) = 1\)  

- Example: Tossing 3 coins (Binomial distribution).


## 2. Probability Density Function (PDF)
- Used for **Continuous Random Variables**.  
- Describes the **likelihood of a value** within an interval.  
- Probability is calculated as the **area under the curve**.  

### Properties
1. \(f(x) \geq 0\)  
2. \(\int_{-\infty}^{\infty} f(x)\,dx = 1\)  

- Example: Normal distribution of heights.


## Key Point
- **PMF** and **PDF** are not separate concepts;  
they are **types of probability distribution functions**.  
- Which one you use depends on whether your random variable is **discrete or continuous**.  


# *Distributions in Statistics*

A **distribution** shows how the values of a random variable are spread (or distributed).  
It tells us the probability of different outcomes.


## Types of Distributions

### 1. Discrete Distributions
Used when the random variable is **discrete** (countable values).  
Examples:
- **Bernoulli Distribution** â†’ Single trial with success/failure.  
- **Binomial Distribution** â†’ Number of successes in fixed trials.  
- **Poisson Distribution** â†’ Counts of events in a fixed interval.



### 2. Continuous Distributions
Used when the random variable is **continuous** (infinite values in an interval).  
Examples:
- **Normal Distribution** â†’ Bell-shaped, most common in nature.  
- **Exponential Distribution** â†’ Time until an event occurs.  
- **Uniform Distribution** â†’ All values in an interval are equally likely.
