# Probability Distributions

*A **Probability Distribution** describes how the probabilities of a random variable are spread over its possible values.  
It tells us **what values the variable can take** and **how likely each value is**.*


- **Random Variable**
    - Value is unknown
    - A function that assigns values to each of experiment outcomes.

**Outsome of any Experiments**

- tossing a coin
- throwing a dice
    - Discrete Outcomes
    - **Probability Mass Functions**

- Probability of height of students below 170cm
- Probability of rain with 7-8 km area.
    - Continuous outcomes
    - **Probability Density Functions**

# Probability Mass Function (PMF)

*Distributions of discrete random variables.*

A **Probability Mass Function (PMF)** is used to describe the probability distribution of a **discrete random variable**.  
It gives the probability that the random variable \(X\) is exactly equal to a certain value \(x\).  


- Example of Probability Mass Function (PMF)

 **Scenario:** Toss a fair coin **2 times**.  
Random variable \(X\) = number of heads.


- Step 1: Sample Space
Possible outcomes = {HH, HT, TH, TT}  

    - HH → 2 heads  
    - HT → 1 head  
    - TH → 1 head  
    - TT → 0 heads  

So, \(X\) can take values {0, 1, 2}.  


- Step 2: Probabilities
Each outcome has probability = \(1/4\).  

    - \(P(X=0)\) = 1/4 = 0.25  
    - \(P(X=1)\) = 2/4 = 0.5  
    - \(P(X=2)\) = 1/4 = 0.25  



- Step 3: PMF Table

| X (Number of Heads) | Probability P(X) |
|----------------------|------------------|
| 0                    | 0.25             |
| 1                    | 0.50             |
| 2                    | 0.25             |

The sum = 0.25 + 0.5 + 0.25 = 1 


# Probability Density Function (PDF)

*Distributions of conntinuous Data*

A **Probability Density Function (PDF)** describes the probability distribution of a **continuous random variable**.  
Since continuous variables can take **infinitely many values**, the probability at an exact point is **0**.  
Instead, probabilities are calculated **over intervals** using the area under the curve.


- A PDF does not give probabilities directly at points.

- Instead, we calculate probabilities using areas under the curve.

# Cumulative Distribution Functions (CDF)

*Summation of all probabilities possible upto a given points*

A **Cumulative Distribution Function (CDF)** gives the probability that a random variable \(X\) is **less than or equal to** some value \(x\).  


Min Value will be zero 

Max value will be 1
