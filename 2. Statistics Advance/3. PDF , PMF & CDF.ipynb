# Relation between PDF, PMF, and CDF

| Concept | Used For | Definition | Formula | Relation with CDF |
|---------|----------|------------|----------|-------------------|
| **PMF (Probability Mass Function)** | Discrete random variables | Gives probability of exact value | \( P(X=x) \) | \( F(x) = P(X \leq x) = \sum_{t \leq x} P(X=t) \) |
| **PDF (Probability Density Function)** | Continuous random variables | Describes density of probability at a point | \( f(x) \geq 0, \int_{-\infty}^{\infty} f(x) dx = 1 \) | \( F(x) = \int_{-\infty}^x f(t) \, dt \) |
| **CDF (Cumulative Distribution Function)** | Both discrete and continuous | Probability that random variable is less than or equal to \(x\) | \( F(x) = P(X \leq x) \) | For discrete: sum of PMF. For continuous: integral of PDF. |

---

**Extra relation:**
- If CDF is differentiable → \( f(x) = \frac{d}{dx}F(x) \) (for continuous case).


# Difference between PDF, PMF, and CDF

| Feature | PDF (Probability Density Function) | PMF (Probability Mass Function) | CDF (Cumulative Distribution Function) |
|---------|-------------------------------------|----------------------------------|-----------------------------------------|
| **Used for** | Continuous random variables | Discrete random variables | Both continuous & discrete |
| **Meaning** | Describes density of probability at a point | Gives exact probability of a value | Gives probability up to a value |
| **Value interpretation** | Value itself is not probability; area under curve gives probability | Value is the actual probability of that point | Value is probability that random variable ≤ x |
| **Range of function** | \( f(x) \geq 0 \), integrates to 1 | \( 0 \leq p(x) \leq 1 \), sums to 1 | \( 0 \leq F(x) \leq 1 \) |
| **Formula** | \( P(a \leq X \leq b) = \int_a^b f(x)dx \) | \( P(X = x) = p(x) \) | \( F(x) = P(X \leq x) \) |
| **Relation** | Derivative of CDF: \( f(x) = F'(x) \) | N/A | Integral/sum of PDF or PMF |
| **Graph shape** | Smooth curve (e.g. bell curve) | Bars for each discrete value | Non-decreasing step (discrete) or S-shape (continuous) |


**PMF** : Probability Mass Function
- Discrete Random Variable
- Ex: Rolling a dice.

**PDF** : Probability density function

- Random Variable is Continuous


## Probability density of pdf plot is the slope (gradient) of CDF at a given point.

### Observations:

Initially in PDF plot, the values are less, so slope of the CDF starts increasing gradually, and the slop is maximum at the highest value of PDF and again starts declining.

- The CDF curve tells you the probability up to a point.
- The PDF tells you how fast that probability is growing at that point.
- So, the slope of the CDF = PDF.
- If the CDF is rising steeply, the PDF is high (more probability packed around there).
- If the CDF is flat, the PDF is near zero (very little probability around there).

    - Example: Standard Normal Distribution
    - PDF (bell curve) → shows where the random variable is most likely.
    - CDF (S-shaped curve) → starts at 0, ends at 1.

- At the center (x=0), the CDF is steepest → slope is big → PDF has its peak there.
- At the tails (far left or right), the CDF flattens → slope is tiny → PDF near 0.
