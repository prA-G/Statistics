# Bernoulli Distributions:

*Any event which have binary operations*

A Bernoulli distribution is a discrete probability distribution of a random variable that has only **two possible outcomes**:
    - Success (1)
    - Failure (0)

- Probability of success = p  
- Probability of failure = 1 - p

- Formula:
    - P(X=1) = p,   P(X=0) = 1-p

### Example: Tossing a Coin
Let's say: 
- Success = Head
- Failure = Tail
    - Probability of Head (p) = 0.5  
    - Probability of Tail (1-p) = 0.5

## Conditions of Bernoulli Distributions:

1. Number of trial should be finite.
2. Each trial should be independent of last trial.
3. Only two possible outcomes.
4. Probability of each outcomes should be same in every trial.

### Examples of Bernoulli Distributions:

1. Tossing a coin
2. Someone has committed a fraud or not.
3. Win or loss
4. Pass or fail
5. Rain or not

### Mean and variance od Bernoulli Distributions:
- Mean (E[X]): The expected value or average outcome of the distribution, which is p. 
- Variance (Var[X]): A measure of how much the outcomes deviate from the mean, calculated as p(1-p). 
- In Summary 
    - Mean: E[X] = p
    - Variance: Var[X] = p(1-p)

# Binomial Distributions

*Any event which has two outcomes*

**In probability theory and statistics, the binomial distribution with parameters n and p is the discrete probability distribution of the number of successes in a sequence of n independent experiments, each asking a yes–no question, and each with its own Boolean-valued outcome: success (with probability p) or failure (with probability q = 1 − p).**
- A single success/failure experiment is also called a Bernoulli trial or Bernoulli experiment 
- A sequence of outcomes is called a Bernoulli process; for a single trial, 
- i.e., n = 1, the binomial distribution is a Bernoulli distribution. 
- The binomial distribution is the basis for the binomial test of statistical significance

**Binomial distribution is the 'n' Bernoulli trial.**

## Mean and Vraince of Binomial Distributions:

*The mean (or expected value) of a binomial distribution is np, and its variance is npq, where 'n' is the number of trials, 'p' is the probability of success, and 'q' is the probability of failure (1-p).*
- Formulas 
    - Mean `(μ): n * p`
    - Variance `(σ²): n * p * q`
Where
    - n: The total number of trials in the experiment. 
    - p: The probability of a single "success" in one trial. 
    - q: The probability of a single "failure" in one trial, which is calculated as 1 - p. 
- How to use them 
    - Identify n, p, and q: from the context of the problem.
    - Calculate the mean: by multiplying 'n' by 'p'.
    - Calculate the variance: by multiplying 'n', 'p', and 'q'.

# Poisson Distributions:

A Poisson distribution is a discrete probability distribution. It gives the probability of an event happening a certain number of times (k) within a given interval of time or space. The Poisson distribution has only one parameter, λ (lambda), which is the mean number of events.

## When to use:
- You can use a Poisson distribution when: 
    - Events are independent of one another.
    - The rate of occurrence is constant.
    - Events are not occurring simultaneously.
    - Events can be counted within a fixed interval.

The Poisson distribution is positively skewed (Skewness > 0) and leptokurtic (Kurtosis > 0), meaning it has a longer tail on the right side and heavier tails than the normal distribution. However, for large values of λ, it becomes increasingly symmetric and bell-shaped, resembling a normal distribution.

### Examples in daily life:
1. No. of people visiting temple/ hospital/ banks/ airport in any hours.
2. No. of accidents every hour at a busy route.
3. No. of emails received every hours.
