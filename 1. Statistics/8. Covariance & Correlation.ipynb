# Covariance


- Covariance shows how two variables change together.
- If both increase together → positive covariance.
- If one increases when the other decreases → negative covariance.
- Value is not standardized, so it depends on the units of variables.

# Correlation


- Correlation is a standardized version of covariance.
- It measures the strength and direction of the linear relationship between two variables.
- Always between -1 and +1.
    - +1 → Perfect positive relationship
    - -1 → Perfect negative relationship
    - 0 → No linear relationship
- Correlation is not depends on slope.

# Advantages and Disadvantages of Covariance and Correlation

## Covariance
### Advantages
- Shows the direction of the linear relationship (positive or negative).
- Useful in portfolio theory (finance) to see how assets move together.
- Simple to compute.

### Disadvantages
- Magnitude is not standardized → depends on units of measurement.
- Hard to interpret strength of relationship (a value of 200 could be weak or strong depending on scale).
- Cannot compare covariance values across different datasets.




## Correlation
### Advantages
- Standardized measure → always between -1 and +1.
- Easy to interpret (closer to +1 or -1 means stronger relationship).
- Unit-free, so can compare across datasets.
- Commonly used in statistics, machine learning, and data analysis.

### Disadvantages
- Only measures linear relationship (ignores non-linear patterns).
- Sensitive to outliers → a few extreme values can distort correlation.
- Does not imply causation (high correlation ≠ one variable causes the other).

# Pearson Correlation Coefficient

## Definition
- Pearson correlation coefficient (often called **r**) measures the strength and direction 
  of the linear relationship between two variables.
- Value ranges from -1 to +1.
- Always measure linear relationship

## Interpretation
- **+1** → Perfect positive linear relationship  
- **-1** → Perfect negative linear relationship  
- **0**  → No linear relationship  

### What to do for understanding non linear relationship?

#### Spearman Rank Correlation 

Spearman rank correlation (ρ or r_s) measures the **strength and direction of the monotonic relationship** between two variables.  
- Instead of raw values, it uses **ranks** of the data.  
- Works for both linear and non-linear monotonic relationships.  
- Value ranges from -1 to +1.
