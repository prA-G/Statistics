# Sampling Techniques

In statistics, analyzing the entire population is often impractical, costly, or time-consuming. Instead, we select a smaller group from the population to study. This process is called sampling.

# Type of Sampling

    - Simple Random Sampling
    - Stratified Sampling
    - Cluster Sampling
    - Systematic Sampling

## 1. Simple Random Sampling

- **Definition:** 
In random sampling, every individual in the population has an equal chance of being selected.

- **How it works:** 
Selection is often done using random number generators, lottery methods, or computer algorithms.

- **Advantages:**
    - Removes bias.
    - Results are more likely to represent the population.

- **Disadvantages:**
    - May still miss important sub-groups.
    - Requires a complete list of the population.
    - Not equal representation across group.

- **Example:** 
Choosing 50 students at random from a university to study their study habits. Every student has an equal chance of selection.

Every member of the population (N) has an equal chance of being selected in the sample.
    - Each person has 1/12 probability of being selected in sampling.


## 2. Stratified Sampling

- **Definition:**
In stratified sampling, the population is divided into subgroups (strata) based on shared characteristics (e.g., age, gender, income level). Then, random samples are drawn from each stratum in proportion to their size.

- **Advantages:**
    - Ensures representation of all sub-groups.
    - More accurate than simple random sampling when population is diverse.

- **Disadvantages:**
    - Requires knowledge of population characteristics.
    - More complex to organize.

- **Example:** 
Suppose a college has 60% male and 40% female students. To survey 100 students:

Randomly pick 60 males and 40 females to maintain the same proportion as the population.

## 3. Cluster Sampling
- **Definition**: 
The population is divided into groups (called *clusters*), and then *entire clusters* are randomly selected.
Then all the individuals in the chosen clusters are selected for the sample.

- **When to use**: 
Useful when the population is large and spread out geographically.

- **Key Point**: 
Clusters should ideally represent the population.

- **Example:**
    - Population = All students in a country.
    - Divide into clusters = Schools.
    - Randomly select 5 schools and survey **all students** in those schools.

## 4. Systematic Sampling

- **Definition**: 
A sample is chosen by selecting every *k-th* element from an ordered list after a random starting point.

- **When to use**: 
Useful when population list is available and evenly distributed.

- **Key Point**: 
Easy and quick, but can introduce bias if theres a hidden pattern.

- **Example:**
    - Population = 1000 customers.
    - Choose every 10th customer after randomly starting from the 4th  (4th, 14th, 24th, ).
