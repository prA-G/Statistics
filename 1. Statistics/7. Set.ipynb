# *SET*

*A **set** is a well-defined collection of distinct objects.  
Objects inside a set are called **elements** or **members**.*

Collection of unordered unique elements.

## Properties of Sets

1. **Well-definedness**  
   - A set must be clear and specific.  
   - Example: {all even numbers less than 10} = {2, 4, 6, 8}  

2. **Distinct Elements**  
   - A set cannot have duplicate elements.  
   - Example: {1, 2, 2, 3} = {1, 2, 3}  

3. **Order Irrelevant**  
   - The order of elements does not matter.  
   - Example: {1, 2, 3} = {3, 2, 1}  

4. **Finite and Infinite Sets**  
   - Finite: limited number of elements → {1, 2, 3, 4}  
   - Infinite: unlimited elements → {1, 2, 3, …}  

5. **Subset**  
   - If every element of set A is in set B, then A is a subset of B.  
   - Example: {1, 2} ⊆ {1, 2, 3, 4}  

6. **Proper Subset**  
   - If A is a subset of B but A ≠ B, then A is a proper subset.  
   - Example: {1, 2} ⊂ {1, 2, 3}  

7. **Universal Set**  
   - A set that contains all elements under consideration.  
   - Usually denoted by \( U \).  

8. **Empty Set (Null Set)**  
   - A set with no elements, denoted by { } or \( \varnothing \).  

9. **Equal Sets**  
   - Two sets are equal if they have exactly the same elements.  
   - Example: {a, b, c} = {c, a, b}  

10. **Power Set**  
    - The set of all subsets of a set.  
    - Example: For A = {1, 2},  
      Power set \( P(A) = \{\varnothing, \{1\}, \{2\}, \{1, 2\}\} \)

11. **Cardinality**  
    - The number of elements in a set.  
    - Example: |{1, 2, 3}| = 3
