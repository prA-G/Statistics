# *Measures of Symmetricity*

*In statistics, **measures of symmetricity** help us understand if a distribution of data is symmetric or skewed around its center.*

It talks about shape of the data.

If anything is exactly towart the left and right is called symmetrical.

# 1. Skewness
- Skewness measures the **degree of asymmetry** in a distribution.
- If data is symmetric → skewness = 0.  
- If data has a long tail on the right → **positive skew**.  
- If data has a long tail on the left → **negative skew**.  

## Types of Skewness
1. **Zero Skewness (Symmetric distribution)**
   - Mean = Median = Mode  
   - Example: Normal distribution  

2. **Positive Skewness (Right-skewed)**
   - Long tail on the right side  
   - Mean > Median > Mode  

3. **Negative Skewness (Left-skewed)**
   - Long tail on the left side  
   - Mean < Median < Mode  

## Interpretation
- **Skewness = 0** → perfectly symmetric distribution  
- **Skewness > 0** → positively skewed (tail to the right)  
- **Skewness < 0** → negatively skewed (tail to the left)  

## Use Case:

- Some of ML algorithms requires symmetric data to build the ML model.
- If not symmetric:
    - Transformation :
        - Log transformation
        - Exponential transformation 
        - Reciprocal transformation
        - Square transformation
        - Square root transformation
        - Box-lox transformation
        - Yeo-Thonson transformation.

### Why data is skewed?

- Nature of data
- Presence of Outliers: To accommodate outliers, the data is skeweed.


# 2. Kurtosis

Kurtosis measures the **peakedness and tail heaviness** of a distribution compared to a normal distribution.


## Types of Kurtosis
1. **Mesokurtic**
   - Kurtosis = 3  
   - Normal distribution (reference point).  

2. **Leptokurtic**
   - Kurtosis > 3  
   - Sharper peak, heavier tails (more outliers).  

3. **Platykurtic**
   - Kurtosis < 3  
   - Flatter peak, lighter tails.  


## Interpretation
- **Kurtosis = 3** → Mesokurtic (normal distribution)  
- **Kurtosis > 3** → Leptokurtic (heavy tails, sharp peak)  
- **Kurtosis < 3** → Platykurtic (light tails, flat peak)
